# Law of Large Numbers

(a)  Write a function `simulateDice(n)` that rolls a six-sided die $n$ times and returns the frequencies of the results, i.e., the number of times 1 appears, the no. of times 2 appears, etc. 

```python
import random

def simulateDice(n):  #function definition
    l=[0,0,0,0,0,0]   #initialising a list of frequencies of 1 to 6
    for i in range(n):  #looping n times
        x=random.randrange(1,7)  #rolling a die, i.e.,generating a random no. between 1 and 6
        l[x-1]+=1    #incrementing the frequency of the rolled no. by 1
    return l  #returning the frequencies

#print(simulateDice(6000000))
```

(c)  Write a function `avgDice(n)` that calls `simulateDice(n)` and compute the average of the $n$ outcomes. Run avgDice(n) 1000 times and plot a bar chart with iterations (i.e., 1 to 1000) against the average values. A sample output looks for $n = 10$ as follows. 


```python
import matplotlib.pyplot as plt
import random

def simulateDice(n):
    l=[0,0,0,0,0,0]
    for i in range(n):
        x=random.randrange(1,7)
        l[x-1]+=1
    return l

def avgDice(n):  
    sum=0  #initialising sum to 0
    l=simulateDice(n)  #calling simulateDice function, passing n as argument
    for i in range(6):  #looping six times
        sum+=(i+1)*l[i]  #after the loop sum will be summation of fi*xi for i = 1 to n
    return sum/n  #returning average as summation of fi*xi for i = 1 to n divided by n
    
l1=[]  #creating an empty list
for i in range(1000):  
    l1.append(avgDice(10))  #appending the average returned by avgDice(n) to the end of the list

plt.bar(range(1,1001),l1)  #creating a bar chart with 1 to 1000 on the x-axis and average of outcomes on the y-axis
plt.xlabel("iteration no.")  #labelling the x-axis
plt.ylabel("average of the outcome")  #labelling the y-axis
plt.title("Distribution of outcomes for n=10")  #giving a title to the plot
plt.show()  #plotting the bar chart
```

# Birthday Paradox

(a) Write a function `simulateBday(n)` that generates $n$ random birthdays (i.e., $n$ random integers between 1 and 366) and determine the number of pairs of same birthdays.

```python
import random
def simulateBday(n):
    l=[0]*366   #creating a list of 366 elements all initialised to 0 representing the no. of birthdays on the day with that index im a year
    num=0  #num is the no. of pairs of birthdays initialised to 0
    for i in range(n):
        x=random.randrange(1,367)  #selecting a random birthday
        if l[x]==0:  #if there is no birthday on that day make it 1
            l[x]=1
        else:  #if there is a birthday on that day make it 0 and increment num by 1
            l[x]=0
            num+=1
    return num #returning no. of pairs of same birthdays
```

(b)   Run `simulateBday(n)` 100 times and determine the number $x$ of times that the $n$ birthdays have an equal pair. Treat $x/n$ as the probability of two same birthdays.

```python
import random
def simulateBday(n):
    l=[0]*366
    
    num=0
    for i in range(n):
        y=random.randrange(1,367)
        if l[y-1]==0:
            l[y-1]=1
        else:
            l[y-1]=0
            num+=1
    return num

n=random.randrange(1,367)  #selecting a random no. of birthdays
x=0  #x is the number of times there is atleast one pair of same birthdays
for i in range(100):
    if simulateBday(n): #will be true if simulateBday(n) is not 0
        x+=1

print(f"number of times that {n} birthdays have an equal pair is {x}")
print(f"probability is {x/100}")
```

# Monty Hall Game

You are on a game show and given the choice of picking one of the three boxes ($X$, $Y$, $Z$). Two of the boxes are empty and one contains a gift. You pick a box, say $X$, and the host (who knows what is inside each of the boxes) opens another box, say $Y$, which is empty. You now have the option of retaining your choice ($X$) or switching it (to $Z$). Is it to your advantage to switch your choice in order to get the gift? This advantage is defined as the fraction of times switching the choice leads to a win when the game is played 1000 times.

```python
import random

x=0 #x is the no. of times switching is advantageous
for i in range(1000):
    gift=random.randrange(3) #gift can be randomly in any one of the three boxes
    player=random.randrange(3) #player chooses one of the three boxes randomly
    host=random.randrange(3)  #host is initialised with a random no.
    while(gift==host or player==host):
        host=random.randrange(3)  
        #host chooses a box which neither contains the gift nor is the one chosen by the player randomly 
        #(in the case that the box chosen by the player and the one containing the gift differ then the 
        #randomness is redundant as there is only one box that the host can choose)
    for j in range(3):
        if(not(j==host or j==player)): #the condition in which j is neither the box chosen by the host 
            #nor the one chosen by the player
            new_choice=j  #the player's new choice after switching will be j
            break #breaking from the loop
    if(new_choice==gift):
        x+=1  #incrementing the no. of times switching is advantageous
print(f"advantage is {x/1000}")
```

# Game of Dice

Simulate the following dice game. 

1. Roll two six-sided dice. 
    * If the sum is 7 or 11 on the first roll, you win. 
    * If the sum is 2, 3 or 12 on the first roll, you lose. 
    * If the sum is 4, 5, 6, 8, 9 or 10 on the first roll, that sum becomes your point. 
2. Continue rolling the dice until the sum is 7 (game lost) or equal to your point (game won). 

If $n$ games are played and $q$ of these games are won, then the chance of winning is $q/n$. Sample simulation of the game for $n=10$ is as follows. What is the chance of winning at this game?

```python
import random
def game():
    x=random.randrange(1,7) #rolling die no.1
    y=random.randrange(1,7) #rolling die no.2
    point=x+y  #sum of the two dice gives the point
    print(f"Player rolled {x} + {y} = {point}")
    print(f"point is {point}")
    if point==7 or point==11: #player wins if sum is 7 or 11 on the first toss
          print("player wins")
          return 1
    if point==2 or point==3 or point==12: #player loses if sum is 2 or 3 or 12 on the first toss
          print("player loses")
          return 0
    while(not (x+y==7 or x+y==point)):  #if game doesn't end on first toss, it continues till the sum is either 7 
                                #in which case the player loses or the same as point in which case the player wins
          x=random.randrange(1,7)
          y=random.randrange(1,7)
          print(f"Player rolled {x} + {y} = {x+y}")
    if(x+y==7):
          print("player loses")
          return 0
    else:
          print("player wins")
          return 1

q=0  #q is the no. of wins
n=10000
for i in range(n):
    if game(): #if game returns 1
        q+=1
print(f"{q} games were won out of {n}.Therefore chances of winning is {q/n}")
```

# Central Limit Theorem 

From a set $S=\{1,1,2,3,5,5,5,7,8,10,12\}$ of numbers, pick $n=1$ numbers at random and compute the average. Repeat this $10^5$ times and plot the frequency of the average values (i.e., how many times each average value occurs)  as a histogram. Repeat this experiment with $n \in \{5,10,30,100,1000\}$.

```python
import random
import matplotlib.pyplot as plt

def avg(n):
    S=[1,1,2,3,5,5,5,7,8,10,12]  #given set S
    sum=0
    for i in range(n):
        sum+=S[random.randrange(0,11)]  #incrementing the sum by a random no. from the set
    return sum/n  #taking average of the values and returning it

l=[1,5,10,30,100,1000]
l1=[]
for j in l:  #traversing l
    for i in range(100000):
        l1.append(avg(j))  #appending l1 with the average returned by avg(n)
    plt.hist(l1,bins=100)  #creating a histogram with 100 small dividisions in the x-axis between starting and ending values
    plt.show()  #plotting a histogram
```
